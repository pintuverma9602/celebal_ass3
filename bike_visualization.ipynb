```python
# Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python

# Load Dataset
df = pd.read_csv("Used_Bikes (1).csv")  
```

```python
# Plot 1: Distribution of Bike Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Bike Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot 2: Average Price by Top 10 Brands
top_brands = df['brand'].value_counts().nlargest(10).index
plt.figure(figsize=(12, 6))
sns.barplot(data=df[df['brand'].isin(top_brands)],
            x='brand', y='price', estimator=np.mean, palette='viridis')
plt.title('Average Bike Price by Top 10 Brands')
plt.ylabel('Average Price')
plt.xlabel('Brand')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 3: Age vs Price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='price', hue='brand', alpha=0.7)
plt.title('Age vs Price of Used Bikes')
plt.xlabel('Age (Years)')
plt.ylabel('Price')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python

# Plot 4: Kilometers Driven vs Price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='kms_driven', y='price', hue='brand', alpha=0.7)
plt.title('Kilometers Driven vs Price')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='owner', order=df['owner'].value_counts().index, palette='Set2')
plt.title('Distribution of Bike Ownership')
plt.xlabel('Ownership Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
top_cities = df['city'].value_counts().nlargest(10).index
plt.figure(figsize=(12, 6))
sns.barplot(data=df[df['city'].isin(top_cities)],
            x='city', y='price', estimator=np.mean, palette='coolwarm')
plt.title('Average Bike Price by Top 10 Cities')
plt.ylabel('Average Price')
plt.xlabel('City')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['price', 'age', 'power', 'kms_driven']].corr(), annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Correlation Between Numeric Features')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='power', y='price', scatter_kws={'alpha':0.5}, line_kws={'color': 'red'})
plt.title('Power vs Price of Bikes')
plt.xlabel('Power (cc)')
plt.ylabel('Price')
plt.show()
```

```python
sns.pairplot(df[['price', 'age', 'power', 'kms_driven']])
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```
